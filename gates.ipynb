This notebook introduces what are called subprograms. This makes the fourth and last control structure that we'll see. The other three are sequence, selection, and loop (repetition).

Sequence just means that statements are run in the order they are coded. An if instruction is a typical selection control. A for is a typical loop control. You need not worry about any details yet as we much of the semester to go.

A subprogram is part of a larger program. In Python, a subprogram -- often referred to as a function or procedure -- is defined with def. In the code cell below, NOT, AND, and OR are all subprograms.

For a function to be used, it needs to be called. You can see this at the end of the code section in the assignment statement F = OR(A,B).

More complex circuits can be made but we do need to be careful about syntax. For example, (A OR B) AND C is coded as AND( OR(A,B), C).

You try one. See if you can write the instruction for (A AND B) OR C when A=0, B=1, and C=1.

```python
def NOT(A):
  if A == 0:
    return 1;
  if A == 1:
    return 0;

def AND(A, B):
  if A == 0 and B == 0:
    return 0;
  if A == 0 and B == 1:
    return 0;
  if A == 1 and B == 0:
    return 0;
  if A == 1 and B == 1:
    return 1;

def OR(A, B):
  if A == 0 and B == 0:
    return 0;
  if A == 0 and B == 1:
    return 1;
  if A == 1 and B == 0:
    return 1;
  if A == 1 and B == 1:
    return 1;

A = int(input('Enter A '))
B = int(input('Enter B '))

F = OR(A,B)
print(F)

A = int(input('Enter A '))
B = int(input('Enter B '))
C = int(input('Enter C '))

F = AND( OR(A,B), C)
print(F)
```
