This program asks for a purchase amount and then takes 8% for a discount. For the purchase amount, enter 120 exactly. Do not put a $ in front of the number. What's output?

```python
discountRate = 0.08

purchase = float(input("Enter the purchase amount: "))

discount = purchase * discountRate
cost = purchase - discount

print("The final cost is " + str(cost))
```
